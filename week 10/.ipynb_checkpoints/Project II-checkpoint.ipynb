```python
# PROJECT I
class OnlineOrangeStore:
    def __init__(self, stock_quantity=100, price_per_orange=1.0):
        self.stock_quantity = stock_quantity
        self.price_per_orange = price_per_orange

    def buy_oranges(self, quantity):
        if quantity <= self.stock_quantity:
            total_cost = quantity * self.price_per_orange
            self.stock_quantity -= quantity
            print("Thank you for your purchase!")
            print("Total cost: $", total_cost)
        else:
            print("Please select a quantity less than or equal to", self.stock_quantity)


store = OnlineOrangeStore()
quantity_to_buy = int(input("Enter the number of oranges to buy: "))
store.buy_oranges(quantity_to_buy)
```

```python
# PROJECT II
class ShapeCalculator:
    def calculate_trapezium_area(self):
        height = float(input("Enter the height of the trapezium: "))
        base1 = float(input("Enter the length of base 1: "))
        base2 = float(input("Enter the length of base 2: "))
        area = (height / 2) * (base1 + base2)
        print("The area of the trapezium is:", area)

    def calculate_rhombus_area(self):
        diagonal1 = float(input("Enter the length of diagonal 1: "))
        diagonal2 = float(input("Enter the length of diagonal 2: "))
        area = 0.5 * diagonal1 * diagonal2
        print("The area of the rhombus is:", area)

    def calculate_parallelogram_area(self):
        base = float(input("Enter the length of the base: "))
        altitude = float(input("Enter the altitude: "))
        area = base * altitude
        print("The area of the parallelogram is:", area)

    def calculate_cube_area(self):
        side_length = float(input("Enter the length of the side of the cube: "))
        area = 6 * (side_length ** 2)
        print("The surface area of the cube is:", area)

    def calculate_cylinder_volume(self):
        radius = float(input("Enter the radius of the cylinder: "))
        height = float(input("Enter the height of the cylinder: "))
        volume = (22/7) * radius ** 2 * height
        print("The volume of the cylinder is:", volume)

calculator = ShapeCalculator()

print("Select an equation to calculate:")
print("1. Area of Trapezium")
print("2. Area of Rhombus")
print("3. Area of Parallelogram")
print("4. Surface Area of Cube")
print("5. Volume of Cylinder")

choice = int(input("Enter your choice (1-5): "))

if choice == 1:
    calculator.calculate_trapezium_area()
elif choice == 2:
    calculator.calculate_rhombus_area()
elif choice == 3:
    calculator.calculate_parallelogram_area()
elif choice == 4:
    calculator.calculate_cube_area()
elif choice == 5:
    calculator.calculate_cylinder_volume()
else:
    print("Please select a number from 1 to 5.")
```

```python
#PROJECT III
import csv

def create_csv(filename, fields, rows):
    with open(filename, 'w', newline='') as csvfile:
        csvwriter = csv.writer(csvfile)
        csvwriter.writerow(fields)
        csvwriter.writerows(rows)

# Create the 'Lagar.csv' file
lagar_fields = ['Lagar']
lagar_rows = [['33 Export'], ['Desperados'], ['Goldberg'], ['Gulder'], ['Heineken'], ['Star']]
lagar_filename = "Lagar.csv"
create_csv(lagar_filename, lagar_fields, lagar_rows)

# Create the 'Stout.csv' file
stout_fields = ['Stout']
stout_rows = [['Legend'], ['Turbo king'], ['Williams'], [''], [''], ['']]
stout_filename = "Stout.csv"
create_csv(stout_filename, stout_fields, stout_rows)

# Create the 'Non-Alcoholic.csv' file
non_alcoholic_fields = ['Non-Alcoholic']
non_alcoholic_rows = [['Maltina'], ['Amstel Malt'], ['Malta Gold'], ['Fayrouz'], [''], ['']]
non_alcoholic_filename = "Non-Alcoholic.csv"
create_csv(non_alcoholic_filename, non_alcoholic_fields, non_alcoholic_rows)

# Ask the user for the file they want to access
print("Select a file to access:")
print("1. Lagar")
print("2. Stout")
print("3. Non-Alcoholic")

# Prompt for numeric input and validate the user's choice
valid_choices = ['1', '2', '3']
choice = input("Enter your choice (1, 2, or 3): ")

if choice not in valid_choices:
    print("Invalid choice.")
    exit()

# Determine the filename based on the user's choice
if choice == '1':
    filename = "Lagar.csv"
elif choice == '2':
    filename = "Stout.csv"
else:
    filename = "Non-Alcoholic.csv"

# Read and print the chosen file
with open(filename, 'r') as csvfile:
    csvreader = csv.reader(csvfile)
    for row in csvreader:
        print(', '.join(row))
```
