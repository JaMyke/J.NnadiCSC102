```python
import csv

class BankApp():
    def __init__(self, username, email_address, pin):
        self.email_address = email_address
        self.pin = pin
        self.balance = 0

    def register(self):
        with open('users.csv', 'a', newline='') as csvfile:
            writer = csv.writer(csvfile)
            writer.writerow([self.email_address, self.pin, self.balance])
            print("Registration successful!")

    def login(self):
        with open('users.csv', 'r') as csvfile:
            reader = csv.reader(csvfile)
            for row in reader:
                if len(row) >= 3 and self.email_address == row[0] and self.pin == row[1]:
                    self.balance = float(row[2])
                    print("Login successful")
                    self.menu()
                    return
            print("Invalid email or PIN.")

    def menu(self):
        print(f"Welcome, {self.email_address}! What would you like to do?")
        print("1. Withdraw")
        print("2. Transfer")
        print("3. Top-up Balance")
        print("4. View Balance")
        choice = int(input("Enter your choice: "))

        if choice == 1:
            amount = float(input("Enter the amount to withdraw: "))
            if amount <= self.balance:
                self.balance -= amount
                print(f"Withdrawal of {amount} successful. Remaining balance: {self.balance}")
            else:
                print("Insufficient funds.")

        elif choice == 2:
            recipient_email = input("Enter the recipient's email: ")
            recipient = None

            with open('users.csv', 'r') as csvfile:
                reader = csv.reader(csvfile)
                for row in reader:
                    if len(row) >= 3 and recipient_email == row[0]:
                        recipient = {
                            'email': row[0],
                            'pin': row[1],
                            'balance': float(row[2])
                        }
                        break

            if recipient:
                amount = float(input("Enter the amount to send: "))
                pin = input("Enter your PIN: ")

                if pin == self.pin:
                    if amount <= self.balance:
                        self.balance -= amount
                        recipient['balance'] += amount
                        print("Transfer successful.")
                    else:
                        print("Insufficient funds.")
                else:
                    print("Incorrect PIN.")
            else:
                print("Invalid recipient.")

        elif choice == 3:
            amount = float(input("Enter the amount to top-up: "))
            self.balance += amount
            print(f"Top-up of {amount} successful. Current balance: {self.balance}")
        elif choice == 4:
            print(f"Current balance: {self.balance}")
        else:
            print("Invalid choice.")

email = input("Enter your email address: ")
pin = input("Enter your PIN: ")

user = BankApp(email, pin)
user.register()
user.login()
```
