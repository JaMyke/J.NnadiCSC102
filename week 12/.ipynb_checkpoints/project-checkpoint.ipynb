```python
#project 1
from abc import ABC, abstractmethod

class coup_de_ecriva(ABC):
    @abstractmethod
    def Fan_Page(self):
        pass

class FC_Cirok(coup_de_ecriva):
    def Fan_Page(self):
        print("Welcome to FC Cirok!")

class Madiba_FC(coup_de_ecriva):
    def Fan_Page(self):
        print("Welcome to Madiba FC!")

class Blue_Jay_FC(coup_de_ecriva):
    def Fan_Page(self):
        print("Welcome to Blue Jay FC!")

class TSG_Walker(coup_de_ecriva):
    def Fan_Page(self):
        print("Welcome to TSG Walker!")

def main():
    club_name = input("Enter the name of the club you support: ")

    if club_name == "FC Cirok":
        club = FC_Cirok()
        club.Fan_Page()
    elif club_name == "Madiba FC":
        club = Madiba_FC()
        club.Fan_Page()
    elif club_name == "Blue Jay FC":
        club = Blue_Jay_FC()
        club.Fan_Page()
    elif club_name == "TSG Walker":
        club = TSG_Walker()
        club.Fan_Page()
    else:
        print("Invalid club name!")

main()
   
```

```python
#project 2
from abc import ABC, abstractmethod

class External_Vendors(ABC):
    @abstractmethod
    def menu(self):
        pass

class Cooperative_Hostel(External_Vendors):
    def menu(self):
        print("Cooperative Cafeteria Menu:")
        print("1. Jollof Rice and Stew - cost #200")
        print("2. White Rice and Stew - cost #200")
        print("3. Fried Rice - cost #200")
        print("4. Salad - cost #100")
        print("5. Plantain - cost #100")

class Faith_Hostel(External_Vendors):
    def menu(self):
        print("Faith Hostel Cafeteria Menu:")
        print("1. Fried Rice - cost #400")
        print("2. White Rice and Stew - cost #400")
        print("3. Jollof Rice - cost #400")
        print("4. Beans - cost #200")
        print("5. Chicken - cost #1000")

class Student_Centre(External_Vendors):
    def menu(self):
        print("Student Centre Cafeteria Menu:")
        print("1. Chicken Fried Rice - cost #800")
        print("2. Pomo Sauce - cost #300")
        print("3. Spaghetti Jollof - cost #500")
        print("4. Amala/Ewedu - cost #500")
        print("5. Semo with Eforiro Soup - cost #500")

def main():
    vendor_name = input("Enter the name of the vendor you are interested in (Cooperative Hostel, Faith Hostel, or Student Center): ")

    if vendor_name == "Cooperative Hostel":
        vendor = Cooperative_Hostel()
    elif vendor_name == "Faith Hostel":
        vendor = Faith_Hostel()
    elif vendor_name == "Student Center":
        vendor = Student_Centre()
    else:
        print("Invalid vendor name!")
main()
```
